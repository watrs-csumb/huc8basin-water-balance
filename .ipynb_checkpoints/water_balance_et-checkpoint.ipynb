```python
import pandas as pd
import geopandas as gpd
import numpy as np
import glob
```

```python
def get_precip_data(huc8id, product):
    '''
    input parameters:
    huc8id: USGS HUC8ID [string or INT]
    product: precipitation data product from data/p/

    output:
    out_df: dataframe of monthly precipitation timeseries for huc8id basin
    '''
    filename = glob.glob(f'data/p/{product}*/*{str(huc8id)}*.csv')[0]
    out_df = pd.read_csv(filename)
    out_df['date']=pd.to_datetime(out_df['date'])
    
    out_df.set_index('date',inplace=True)
    
    return out_df

def get_et_data(huc8id, model):
    '''
    input parameters:
    huc8id: USGS HUC8ID [string or INT]
    model: OpenET model name data product from data/et/

    output:
    out_df: dataframe of monthly evapotranspiration timeseries for huc8id basin
    '''
    # model='ensemble'
    # huc8id=17010202
    print(huc8id,17010202)
    print(str(huc8id))
    print(str(huc8id)==str(17010202))
    
    # filename = glob.glob(f'data/et/{model}*/*{huc8id}*.csv')[0]
    print(glob.glob(f'data/et/{model}*/huc_{17010202}_*.csv'))
    print(glob.glob(f'data/et/{model}*/huc_{huc8id}_*.csv'))
    # out_df = pd.read_csv(filename)
    # out_df['date']=pd.to_datetime(out_df['date'])
    # out_df.set_index('date',inplace=True)
    # return out_df
    
# data/et/ensemble_monthly_et/huc_17010202_ensemble_monthly_et.csv
```
